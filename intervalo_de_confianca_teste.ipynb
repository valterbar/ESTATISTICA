```python
import matplotlib.pyplot as plt
import numpy as np
from scipy import stats

# Dados da amostra
media = 10
desvio_padrao = 2
tamanho_amostra = 30

# Gerando uma amostra aleatória de uma distribuição normal
amostra = np.random.normal(loc=media, scale=desvio_padrao, size=tamanho_amostra)

# Intervalo de confiança de 95%
limite_inferior, limite_superior = stats.norm.interval(0.975, loc=media, scale=desvio_padrao / math.sqrt(tamanho_amostra))

# Plotando o histograma
plt.hist(amostra, bins=10, density=True, alpha=0.6, color='g')

# Plotando o intervalo de confiança
plt.axvspan(limite_inferior, limite_superior, color='gray', alpha=0.25)

# Adicionando título e rótulos
plt.title('Histograma com Intervalo de Confiança de 97,5%')
plt.xlabel('Valores')
plt.ylabel('Frequência')

# Mostrando o gráfico
plt.show()
```
